Reproducing figures of the paper:

Spence, D. A., and D. L. Turcotte. “Magma-Driven Propagation of Cracks.” Journal of Geophysical Research: Solid Earth 90, no. B1 (1985): 575–80. https://doi.org/10.1029/JB090iB01p00575.

Eq. 31:

$$ l = 0.7756 A^{1/2} \left( \frac{\mu}{\eta (1-\nu)} \right) t^{2/3} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.logspace(1, 6, 50)
```

```python
μ = 2e10 #Pa
ν = 0.25
```

```python
A_ = np.array([1e-2, 1, 1e2]).T #m^2/s
η_ = np.array([0.1, 1e2])
```

```python
plt.figure()
linestyle = ["-", "--"]
for A in A_:
    for j, η in enumerate(η_):
        l = 0.7756 * A**0.5 * (μ/(η*(1-ν)))**(1.0/6) * t**(2.0/3)
        plt.plot(t, l, ls=linestyle[j], label=f"A = {A}, η = {η}")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.xlabel("t (s)")
plt.ylabel("l (m)")
plt.show()
```

```python
plt.figure()
plt.plot(t, l)
plt.show()
```

```python
plt.figure()
linestyle = ["-", "--"]
for A in A_:
    for j, η in enumerate(η_):
        l = 0.7756 * A**0.5 * (μ/(η*(1-ν)))**(1.0/6) * 2.0/3*t**(2.0/3-1)
        plt.plot(t, l, ls=linestyle[j], label=f"A = {A}, η = {η}")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.xlabel("t (s)")
plt.ylabel("v (m/s)")
plt.show()
```
