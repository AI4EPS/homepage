```python
import obspy
from obspy.clients.fdsn.client import Client
```

```python
def print_event(event):
    event.plot("ortho")
    print("-------------------------------------")
    print(event[0])
    print("-------------------------------------")
    print(event[0].origins[0])
    print("-------------------------------------")
    print(event[0].magnitudes[0])
    print("-------------------------------------")
    try:
        print(event[0].amplitudes[0])
    except:
        pass
    print(event[0].picks[0])
    print("-------------------------------------")
    print(event[0].origins[0].arrivals[0])
    print("=====================================")

    
```

```python
client = Client("USGS")
event = client.get_events(eventid="ak01862cwg6j", includeallorigins=True)
print_event(event)
```

```python
client = Client("USGS")
event = client.get_events(eventid="ci40403984", includeallorigins=True)
print_event(event)
```

```python
client = Client("USGS")
event = client.get_events(eventid="nc73836266", includeallorigins=True)
print_event(event)
```

```python
client = Client("NCEDC")
event = client.get_events(eventid="73836266", includearrivals=True, includeallmagnitudes=True)
print_event(event)
```
